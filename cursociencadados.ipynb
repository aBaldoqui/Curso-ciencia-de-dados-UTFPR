```python
!pip install pandas numpy plotly
import pandas as pd
import numpy as np
import plotly.graph_objects as go
```

```python
file1 =  '/gdp_data.csv'

data = pd.read_csv(file1, sep = ',', index_col = None)
data
```

```python
file2 = '/country_codes.csv'

country = pd.read_csv(file2, sep = ',', index_col =None)
country
```

```python
df = pd.merge(data, country, on='country_code', how = 'inner')
df
```

```python
data.info()
```

```python
df['year'] = df['year'].astype(str)
data.info()
```

```python
print(df['value'].min())
print(df['value'].quantile(0.25))
print(df['value'].median())
print(df['value'].quantile(0.75))
print(df['value'].max())
print(df['value'].mean())
```

```python
brasildata = df['country_name'] == 'Brazil' #isso é uma mascara
brasilmd = df.loc[brasildata, ['value']].mean()
brasilmd
```

```python
# outra forma
brasildata = df.loc[brasildata]
brasildata
```

```python
brm75 = (df['country_name'] == 'Brazil') & (df['value']> data['value'].quantile(0.75)) #isso é uma mascara mais complicada
brasildata.loc[brm75]
```

```python
df.loc[df['year'] == '2022']
```

```python
fig = go.Figure()
fig.add_trace(go.Bar(x=brasildata['year'], y=brasildata['value'], name = 'PIB Brasileiro', marker_color = 'green'))


fig.update_layout(
    title = "Evolução do PIB Brasileiro",
    xaxis = {
        'title' : "Ano"
    },
    yaxis = {
        'title' : "valor"
    }
)

fig.show()
```

```python
m = data.groupby('year')['value'].mean().reset_index()
```

```python
fig = go.Figure()
fig.add_trace(go.Bar(x=brasildata['year'], y=brasildata['value'], name = 'PIB Brasileiro', marker_color = 'green'))

fig.add_trace(go.Scatter(x=m['year'], y=m['value'], name = 'Media do PIB mundial',marker_color = 'red', mode = 'lines'))

fig.update_layout(
    title = "PIB Brasileiro x Media PIB Mundial",
    xaxis = {
        'title' : "Ano"
    },
    yaxis = {
        'title' : "valor"
    }
)

fig.show()
```
